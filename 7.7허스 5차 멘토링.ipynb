객체지향을 하는 이유

>같은 목적과 기능을 위해 객체로 묶인 코드 요소들은 객체 내부에서만 강한 응집력을 발휘하고 객체 외부에 주는 영향은 줄이게 된다.즉 코드수정에 대한 부작용을 줄일 수 있다.



```python
#객체지향 클래스 예제
class Car:
  def __init__(self):
    self.color = 0xFF0000 #바디의 색 (여기서는 10진수로 색깔 표현)
    self.wheel_size = 16 #바퀴의 크기
    self.displacement = 2000 #엔진 배기량

#car 클래스 안에 전진 후진 좌회전,우회전 메소드를 정의
  def forward(self):
    pass
  def backward(self):
    pass
  def turn_left(self):
    pass
  def turn_right(self):
    pass
  
#car 클래스 정의 종료. 아래는 car 클래스의 인스턴스를 정의하고 사용하는 코드 

#my_car의 정보를 출력
if __name__ == '__main__':
  my_car = Car()
  print('0x{:02X}'.format(my_car.color))
  print(my_car.wheel_size)
  print(my_car.displacement)

#my_car의 메소드를 호출
  my_car.forward()
  my_car.backward()
  my_car.turn_left()
  my_car.turn_right()
```

```python
 class InstanceVar:
   def __init__(self):
     self.text_list = []  #__init__()메소드 안에서 test_list를 정의
   def add(self, text):
     self.text_list.append(text) #self가 가리키는 자신은 메소드가 소속되어 있는 객체를 가리킨다
   def print_list(self):
     print(self.text_list)
if __name__ == '__main__':
  a = InstanceVar()
  a.add('a')
  a.print_list() # ['a'] 출력을 기대

  b = InstanceVar()
  b.add('b')
  b.print_list() # ['b'] 출력을 기대
```

```python
#클래스 안에 정의되었지만 self를 매개변수로 받지 않는다.
class Calculator:
  @staticmethod #데코레이터로 수식하며, self 매개변수 없이 정의한다.
  def plus(a,b):
    return a+b
    
  @staticmethod
  def minus(a,b):
    return a-b

  @staticmethod
  def multiply(a,b):
    return a*b

  @staticmethod
  def divide(a,b):
    return a/b

#정적 메소드는 인스턴스 대신 클래스 이름을 통해 메소드에 접근합니다. "클래스이름.메소드()"꼴
if __name__ == '__main__':
  print("{0} + {1} = {2}".format(7,4, Calculator.plus(7,4)))
  print("{0} - {1} = {2}".format(7,4, Calculator.minus(7,4)))
  print("{0} * {1} = {2}".format(7,4, Calculator.multiply(7,4)))
  print("{0} / {1} = {2}".format(7,4, Calculator.divide(7,4)))
```

```python
class A:
  def __init__(self):
    self.message = "Hello"

class B(A): #A클래스 상속 ==> 다중상속 가능(B클래스안에 여러클래스가 들어가 있는것)
  def __init__(self):
    print("B.__init__()")

    super().__init__() #A.__init__(self)와 동일한 결과
    print("self.message is " +self.message)

if __name__ == "__main__":
  b = B()     
```

```python
#오버라이딩 
class A:
  def method(self):
    print("A")

class B(A):
  def method(self):
    print("B")

class C(A):
  def method(self):
    print("C")

#상속했음에도 자신만의 method()를 새로 정의했다 == 오버라이딩
A().method()
B().method()
C().method()
```
