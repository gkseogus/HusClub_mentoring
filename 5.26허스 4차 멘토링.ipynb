## 객체와 클래스


*   객체 지향 프로그래밍
*   객체(Object) = 속성(Attribute)+기능(Method)

*   클래스와 객체의 관계는 아래와 같다.
*   클래스: 자료형, 객체: 변수

*   클래스도 코드블록을 가지므로 '안', '밖'의 개념이 존재

*   클래스 안에서만 접근이 가능한 멤버: Private

*   클래스 안과 밖에서 접근 가능한 멤버: Public










```python
class Car:  #car 클래스 정의 
  def __init__(self):  #__init__ 메소드를 이용한 초기화
    self.color = 'blue' #바디의 색 정의
    self.wheel_size = 16  #바퀴의 크기 정의
    self.displacement = 2000 #엔진 배기량
  
  def forward(self): #전진 함수 정의
    pass
  
  def backward(self): #후진 함수 정의
    pass
    
  def turn_left(self): #좌회전 함수 정의
    pass
    
  def turn_right(self): #우회전 함수 정의
    pass
  
if __name__ == '__main__': #__name__, __main__ 메소드를 이용
  my_car = Car()  #앞에서 정의한 Car 클래스의 객체 my_car 생성

  #my_car 의 정보를 출력
  print('{0}'.format(my_car.color))  
  print(my_car.wheel_size)
  print(my_car.displacement)

  #my_car 메소드를 호출
  my_car.forward()
  my_car.backward()
  my_car.turn_left()
  my_car.turn_right()
```

```python
#Calculator이란 클래스 정의
#정적 메소드는 @staticmethod 데코레이터로 수식한다. self 매개변수 없이 정의
class Calculator:
  @staticmethod
  def lpus(a,b):
    return a+b
  
  @staticmethod
  def lpus(a,b):
    return a-b
  
  @staticmethod
  def lpus(a,b):
    return a*b
  
  @staticmethod
  def lpus(a,b):
    return a/b 
  
#정적 메소드는 인스턴스 대신 클래스 이름을 통해 메소드에 접근
#"클래스이름.메소드()"의 꼴
if __name__ == '__main__': 
  print("{0} + {1} = {2}">format(7,4,Calculator.plus(7,4)))
  print("{0} - {1} = {2}">format(7,4,Calculator.minus(7,4)))
  print("{0} * {1} = {2}">format(7,4,Calculator.multiply(7,4)))
  print("{0} / {1} = {2}">format(7,4,Calculator.divide(7,4)))
```

```python
class HasPrivate:
  def __init__(self):
    self.public = "Public."
    self.__private = "Private."
  
  def print_from_internal(self):
    print(self.public)
    print(self.__private)

obj = HasPrivate()
obj.print_from_internal()
```

## 오버라이딩(Overriding)

*   부모 클래스로부터 상속받은 메소드를 다시 정의하다 라는 뜻



```python
class A:
  def method(self):
    print('A')
class B(A): #클래스 A를 상속 받음
  def method(self): #하지만 자신만의 method()를 새로 정의해 A의 클래스를 상속받지 않음 == 오버라이딩
    print('B')
class C(A):
  def method(self):
    print('C')

A().method()

B().method()

C().method()
```
