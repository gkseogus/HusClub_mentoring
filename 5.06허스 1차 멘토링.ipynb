### ***딕셔너리***

 > := 딕셔너리는 리스트와 다르게 변경이 불가능한 형식(문자열,숫자열...등등)이면 어떤 자료형든 사용 가능



```python
dic = {}
dic['파이썬'] = 'www.python.org' #파이썬 == key
dic['마이크로소프트'] = 'www.microsoft.com'
dic['애플'] = 'www.apple.com'
print(dic['파이썬'])
type(dic) #문자열의 종류를 알려주는 함수
```

```python
dic['파이썬']
```

```python
dic.keys()
```

```python
dic.items()
```

```python
print('애플' in dic.keys())
print('사과' in dic.keys())
```

```python
dic = { '파이썬': 'www.python.org', '마이크로소프트': 'www.microsoft.com'}
dic.pop('애플')
dic
```
