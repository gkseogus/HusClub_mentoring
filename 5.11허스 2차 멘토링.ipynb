if문 예제

```python
print('요일(월~일)을 입력하세요 :')
dow = input() #dow에 입력 값을 넣어준다

if dow == '월':  #dow에 입력된 값이 월요일 이라면
  print('Monday')
elif dow == '화':
  print('Tuesday')
elif dow == '수':
  print('Wednesday')
elif dow == '목':
  print('Thursday')
elif dow == '금':
  print('Friday')
elif dow == '토':
  print('Saturday')
elif dow == '일':
  print('Sunday')
else:
  print('잘못 입력된 요일입니다.')
```

```python
print('수를 입력하세요:')
a = int (input())

if a>10 and a%2==0:   #짝수는 2로 나누었을때 나머지가 0이면 짝수라고 한다
  print('10보다 큰 짝수')
elif a>10 and a % 2 != 0 :  #반대로 홀수는 나머지가 0이 아니면 홀수.
  print('10보다 큰 홀수')
elif a % 2 == 0:
  print('10 이하의 짝수')
else : 
  print('10 이하의 홀수') 
```

while문 예제

```python
print('몇 번 반복할까요?:')
limit = int(input())
count = 0                         #0으로 초기화를 해줌으로써 0부터 시작
while count < limit:              #count가 limit보다 작은 동안은 계속해서 while문 실행
  count = count +1                #while문이 반복될 동안 count는 1씩 증가
  print('{0}회 반복.'.format(count)) #format 함수를 사용해 {}위치에 count 데이터를 넣는다
print('반복이 종료되었습니다.')
```

for문 예제

```python
for i in range(1,6):    #for문에서 i는 1부터 6까지 실행  
  for j in range(i):    #바깥 for문에서 입력한 반복변수 i가 멈춘 값
    print("*",end = "") #end=""을 매개변수로 입력하면 줄바꿈을 출력하지 않는다.
  print()               #아무것도 입력하지 않으면 줄바꿈만 출력
```

continue & break의 예제

```python
for i in range(10):
  if i % 2 == 1:
    continue      #i가 홀수일 때 코드블록의 나머지 부분을 실행하지 않고 다음 반복으로 바로 건너간다
  print(i)        #continue가 실행중일때는 이 코드는 실행하지 않음
```

```python
i = 0                 #i를 0으로 초기화해서 0부터 시작
while(True):          #while 실행문
  i = i+1             #while문이 진행되는 동안 i는 1씩 증가
  if i ==100:         
    print('i가 {0}이 됐습니다. 반복문을 중단합니다.'.format(i))   
    break   #i 가 100이면 break해 while문을 강제로 종료 
  print(i)
```
