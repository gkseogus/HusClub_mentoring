## ***1.함수 정의하기***


>함수는 def 키워드를 이용해서 코드블록에 이름을 붙인 형태이다.

> 문법: def 함수이름(매개변수 목록):
  

> #코드블록


 

>   return 결과









```python
def my_abs(arg): #arg란 매개변수 목록// my_abs란 함수명
  if(arg<0):
    result = arg*-1 
  else:
    result = arg
  return result   #result를 return해 함수 반환
my_abs(-5)
```

```python
def print_team(**players): #매개변수 앞에 **를 붙이면 딕셔너리 가변매개변수
  for k in players.keys():  #매개변수 앞에 *를 붙이면 튜플 가변매개변수
    print('{0}={1}'.format(k,players[k]))
    
print_team(카시야스='GK',호날두='FW',알론소='MF',페페='DF')  
```

```python
#재귀함수
def factorial(n):
  if n == 0:
    return 1
  elif n>0:
    return factorial(n-1)*n
factorial(5)
```

```python
#중첩함수
import math
def stddev(*args): #표준편차
  def mean(): #평균
    return sum(args)/len(args)
  def variance(m): #분산
    total = 0
    for arg in args:
      total += (arg -m) **2
    return total/(len(args)-1)
  v = variance(mean())
  return math.sqrt(v)
stddev(2.3,1.7,1.4,0.7,1.9)
```
